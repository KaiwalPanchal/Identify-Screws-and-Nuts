```python
#V channel
import cv2
import os

# Define the folder paths
input_folder = 'ScrewAndBolt_20240713'
output_folder = 'v_channel'

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop through all the files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.jpg') or filename.endswith('.png'):  # Add other image formats if needed
        # Read the image
        img = cv2.imread(os.path.join(input_folder, filename))
        
        # Check if image is successfully loaded
        if img is None:
            print(f"Failed to load image {filename}")
            continue
        
        # Convert the image from BGR to HSV
        hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
        
        # Extract the V channel
        v_channel = hsv_img[:, :, 2]
        
        # Save the V channel image to the output folder
        cv2.imwrite(os.path.join(output_folder, filename), v_channel)

print("V channel extraction complete.")
```

```python
#OTSU
# Define the folder paths
input_folder = 'v_channel'
output_folder = 'otsu'

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop through all the files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.jpg') or filename.endswith('.png'):  # Add other image formats if needed
        # Read the image in grayscale
        img = cv2.imread(os.path.join(input_folder, filename), cv2.IMREAD_GRAYSCALE)
        
        # Check if image is successfully loaded
        if img is None:
            print(f"Failed to load image {filename}")
            continue
        
        # Apply Otsu's thresholding
        _, otsu_img = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
        
        # Save the thresholded image to the output folder
        cv2.imwrite(os.path.join(output_folder, filename), otsu_img)

print("Otsu's thresholding complete.")
```

```python
#Dist
# Define the folder paths
input_folder = 'OTSU'
output_folder = 'Dist'

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop through all the files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.jpg') or filename.endswith('.png'):  # Add other image formats if needed
        # Read the image
        img = cv2.imread(os.path.join(input_folder, filename), cv2.IMREAD_GRAYSCALE)
        
        # Check if image is successfully loaded
        if img is None:
            print(f"Failed to load image {filename}")
            continue
        
        # Apply binary thresholding
        _, binary_img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

        # Apply distance transform
        dist_transform = cv2.distanceTransform(binary_img, cv2.DIST_L2, 5)
        
        # Normalize the distance transform image to fit within [0, 255]
        dist_transform = cv2.normalize(dist_transform, None, 0, 255, cv2.NORM_MINMAX)
        dist_transform = dist_transform.astype('uint8')

        # Save the distance transformed image to the output folder
        cv2.imwrite(os.path.join(output_folder, filename), dist_transform)

print("Distance transform complete.")
```

```python
#Contours
# Define the folder paths
input_folder = 'invert'
output_folder = 'final_result2'

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop through all the files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.jpg') or filename.endswith('.png'):  # Add other image formats if needed
        # Load the image
        image_path = os.path.join(input_folder, filename)
        img = cv2.imread(image_path)

        # Check if image is successfully loaded
        if img is None:
            print(f"Failed to load image {filename}")
            continue

        # Convert the image to grayscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Apply binary thresholding
        _, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)

        # Find contours on the thresholded image
        contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

        # Draw the bounding rectangles on the original image for filtered contours
        image_with_rectangles = img.copy()
        for i in range(len(contours)):
            # Check if the contour has no children (hierarchy[0][i][2] == -1)
            if hierarchy[0][i][2] == -1:
                x, y, w, h = cv2.boundingRect(contours[i])
                
                # Check width and height limits
                if w <= 250 and h <= 250:
                    cv2.rectangle(image_with_rectangles, (x, y), (x + w, y + h), (0, 255, 0), 2)

        # Save the processed image to the output folder
        output_path = os.path.join(output_folder, filename)
        cv2.imwrite(output_path, image_with_rectangles)
        print(f"Bounding rectangle drawing complete for {filename}. Image saved at {output_path}.")

print("Processing of all images complete.")
```
